```python
# There can use the Kolmogorov-Smirnov test, which access the goodness of fit of the distibution with respect to the data. This test is based on a distance between empirical distribution function of the data and the CDF of the reference distribution
# The p-value is very low in the exponential distribution which can be seen as the null hypothesis (stating that the observed data stems from an exponential distribution with a maximum likelihood rate parameter) and can be rejected with high confidence. Hence, there will try another distribution which is Birnbaum-Sanders distribution, it will be used in the model failure times as well.
st.kstest(spendingScore, dist.cdf, args)
```

```python
# Import all the library which will be used later
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
# This is used to build the Model
# There is using the KMeans Algorithm to decide the optimum cluster number , and the KMeans++ is using Elbow Mmethod
# The way to find out the K for KMeans, I use the Elbow Method which is already implemented in the KMeans++ to do the calculation.
from sklearn.cluster import KMeans
wcss=[]
df1 = new_df[['Age', 'spendingScore']].values
```

```python
# In the normal case, there usually we assume the maximum number of cluster be 10 
# There can use the averaging to judge the number of clusters
# There is used to get the maximum number of cluster

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
```

```python
# This is using to view the ELBOW method to get the optimal value of K 
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# This curve has shown that the last elbow comes at k=5, so the number of clusters become 5
#Model Build
kmeansmodel = KMeans(n_clusters= 5, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(df1)
```

```python
# That is used to view all the clusters 
plt.scatter(df1[y_kmeans == 0, 0], df1[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(df1[y_kmeans == 1, 0], df1[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(df1[y_kmeans == 2, 0], df1[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(df1[y_kmeans == 3, 0], df1[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(df1[y_kmeans == 4, 0], df1[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
# Import library
from sklearn.model_selection import learning_curve

train_size, train_score, test_score = \
        learning_curve(svr, x, y, train_sizes=np.linspace(0.1, 1.0, 10), cv=5, shuffle=True)
```

```python
# Define all the training data and testing data for getting their mean, standard deviation
train_mean = np.mean(train_score, axis=1)
train_std = np.std(train_score, axis=1)
test_mean = np.mean(test_score, axis=1)
test_std = np.std(test_score, axis=1)
```

```python
# Plot the graph by using
plt.plot(train_size, train_mean, color='blue', marker='o', markersize=5)
plt.fill_between(train_size, train_mean+train_std, train_mean-train_std, alpha=0.15, color='blue')
plt.plot(train_size, test_mean, color="red", marker='s')
plt.fill_between(train_size, test_mean+test_std, test_mean-test_std, alpha=0.15, color='red')
```
